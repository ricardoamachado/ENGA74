Utilizando o Python, gere 100.000 realizações de uma variável aleatória Gaussiana (X) de média 10 e desvio padrão 3, para essa variável aleatória.

### Plote o histograma.

```python
media = 10
desv_pad = 3
N = 100000
np.random.seed(50)
distribuicao = np.random.normal(loc=media, scale=desv_pad, size=N)
```

```python
plot_historam = sns.histplot(distribuicao)
plt.xlabel("Valor da variável aleatória")
plt.ylabel("Frequência absoluta")
```

### Plote o boxplot.

```python
plot_boxplot = sns.boxplot(y = distribuicao, width= 0.5)
plt.ylabel("Valor da variável aleatória")
```

### Plote o violinplot.

```python
sns.violinplot(distribuicao)
plt.ylabel("Valor da variável aleatória")
```

### Plote uma estimativa da função densidade de probabilidade (pdf).

```python
sns.kdeplot(distribuicao,color='red')
sns.histplot(distribuicao,stat='density',color='blue')
```

### Estime os valores da média, desvio padrão, IQR e compare com os valores teóricos esperados.

```python
print(f"Média: {np.mean(distribuicao):.5f}\nDesvio Padrão: {np.std(distribuicao,ddof=1):.5f}")
```

Como esperado, tem-se $\mu \approx 10$ e $\sigma \approx 3$ para a distribuição.

```python
print(f"Quartil 25%: {np.percentile(distribuicao, 25):.5f}")
print(f"Quartil 75%: {np.percentile(distribuicao, 75):.5f}")
print(f"Intervalo interquartil: {iqr(distribuicao):.5f}")
```

## Questão 2:

Some a variável aleatória $X$ criada na Questão I às variáveis $Y_i$ definidas a seguir e repita as análises dos itens a, b, c, d da Questão I para a nova variável $Z_i = X + Y_i$

```python
np.random.seed(50)
dist_y1 = np.random.uniform(-0.5,0.5,size=N)
dist_y2 = np.random.uniform(0,1,size=N)
dist_y3 = np.random.normal(loc=0,scale=0.5,size=N)
dist_y4 = np.random.normal(loc=3,scale=0.5,size=N)

dist_z1 = distribuicao + dist_y1
dist_z2 = distribuicao + dist_y2
dist_z3 = distribuicao + dist_y3
dist_z4 = distribuicao + dist_y4
```

### 2 a) Distribuição $Z_1$

#### Plote o histograma.

```python
plot_historam = sns.histplot(dist_z1)
plt.xlabel("Valor da variável aleatória")
plt.ylabel("Frequência absoluta")
```

#### Plote o boxplot.

```python
plot_boxplot = sns.boxplot(y = dist_z1, width= 0.5)
plt.ylabel("Valor da variável aleatória")
```

#### Plote o violinplot.

```python
sns.violinplot(dist_z1)
plt.ylabel("Valor da variável aleatória")
```

#### Plote uma estimativa da função densidade de probabilidade (pdf).

```python
sns.kdeplot(dist_z1,color='red')
sns.histplot(dist_z1,stat='density',color='blue')
```

#### Estime os valores da média, desvio padrão, IQR e compare com os valores teóricos esperados.

```python
print(f"Média: {np.mean(dist_z1):.5f}. Desvio Padrão: {np.std(dist_z1,ddof=1):.5f}")
```

A média da soma de duas variáveis aleatórias independentes é dada por: $\mu_{(x+y)} = \mu_x + \mu_y$

A variância de duas variáveis aleatórias independentes é: $\sigma_{(x+y)}^2 = \sigma_x^2 + \sigma_y^2$


Para a variável aleatória $X$, tem-se $\mu_x = 10$ e $\sigma_x = 3$.

Para a variável aleatória $Y_1$, tem-se $\mu_y = \frac{0,5 - 0,5}{2} = 0$ e $\sigma_y^2 = \frac{(0,5-(0,5))^2}{12} = \frac{1}{12}$.

Dessa forma, $\mu_{z_1} = 10$ e $\sigma_{z_1} = \sqrt{3^2 + \frac{1}{12}^2} = 3,00116$. Portanto, os valores obtidos de forma experimental são coerentes com o esperado.

```python
print(f"Quartil 25%: {np.percentile(dist_z1, 25):.5f}")
print(f"Quartil 75%: {np.percentile(dist_z1, 75):.5f}")
print(f"Intervalo interquartil: {iqr(dist_z1):.5f}")
```
